### Реализация задания 3.3 для функции $y(x) = sin(x)^4$

Перед началом анимации графическая область должна содержать следующие графические объекты:

Начальное состояние графической области реализуем с помощью пользовательской функции init.

### Задание 3.4. Движение секущей прямой к заданной линии на плоскости (версия 2)

Создаййте анимацию движения точки по графику функции $x = x(t), y = y(t)$ по значениям $t$, последовательно изменяющимся от $t_{min}$ до $t_{max}$ с шагом $step$. Функция $x = x(t), y = y(t)$ и значения переменных $t_{min}, t_{max}, step$ задаются такими же, как и при выполнении Задания 3.1b.

```python
fig4 = plt.figure()
ax4 = plt.axes()
plt.axis([x0_min-1, x0_max+1, y0_min-1, y0_max+1])
def init():
    global point_B, secant_line
    curve, = ax4.plot(x0, y0,'b')
    A = np.array([x0[0], y0[0]])
    point_A, = ax4.plot(A[0],A[1],'ro')
    B = np.array([x0[-1], y0[-1]])
    point_B, = ax4.plot(B[0],B[1],'go')
    secant_p = [A + (B - A)*t for t in [-2, 2]]
    secant_p = np.array(secant_p)
    secant_line, = ax4.plot(secant_p[:,0], secant_p[:,1],'g')
    plt.legend([r'$x=\ln\sqrt{\frac{1-t}{1+t}}\quad y=\sqrt{1-t^2}$','Unmovable point A','Мovable point B','Sec'])
def at_frame3(point):
    global point_B, secant_line
    A = np.array([x0[0], y0[0]])
    B = np.array([point[0],point[1]])
    point_B.set_data([point[0]],[point[1]])
    if B[0]==A[0]:
        t = sympy.Symbol('t')
        x_d=(3*t**2+1)/(3*t**3+1)
        y_d = sympy.sin(t**3/3+t)
        x_difx=sympy.diff(x_d,t)
        y_difx = sympy.diff(y_d, t) 
        k=y_difx.subs(t, A[0])/x_difx.subs(t,A[0])
        y_kasstart = np.array([A[0],A[1]])
        y_kasend = np.array([A[0]+1,k+A[1]])
        y_kas=[y_kasstart+ (y_kasend-y_kasstart)*t for t in [-10, 10]]
        y_kas = np.array(y_kas)
        secant_line.set_data(y_kas[:,0], y_kas[:,1])
        secant_line.set_color('r')
    else:
        secant_p = [A + (B - A)*l for l in [-10, 10]] 
        secant_p = np.array(secant_p)
        secant_line.set_data(secant_p[:,0], secant_p[:,1])
FuncAnimation(fig4, at_frame3, frames=matrix0[::-1], init_func=init, repeat=False, interval=40)
```
