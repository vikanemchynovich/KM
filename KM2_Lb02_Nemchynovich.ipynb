Все атрибуты объекта составляют пространство имен объекта.


Доступ к атрибуту attr объекта obj производится с применением синтаксиса

Атрибуты объектов могут быть именами переменных или именами функций. Если

алфавитно-цифровые, а специальные символы.


Коллекции — это наборы объектов, которые делятся на три вида:
Последовательности — это позиционно упорядоченные коллекции. Встроенные

Отображения — это неупорядоченные коллекции объектов, хранящихся по

Множества — это неупорядоченные коллекции не повторяющихся объектов.

Выражение (expression) в Python — это конструкция языка в виде комбинации
